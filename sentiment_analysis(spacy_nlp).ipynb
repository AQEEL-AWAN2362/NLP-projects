```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Emotion_classify_Data.csv")
df.shape
```

```python
df.head(5)
```

```python
#checking the distribution of Emotion
df.Emotion.value_counts()
```

```python
# encoding the target variable
df["emotion_num"]=df['Emotion'].map(
    {"joy":0,
    "fear":1,
    "anger":2})

df.head(5)
```

```python
# train test splitt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    df.Comment,
    df.emotion_num,
    test_size=0.2,
    stratify=df.emotion_num) # for equal distribution of sample
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# making the pipeline
clf=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("mnnb",MultinomialNB())
])

clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
# making the pipeline
clf=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("rfc",RandomForestClassifier())
])

clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
# making the pipeline
clf=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("svc",SVC())
])

clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
clf.score(x_test,y_test)
```

```python
# preprocess tha text
import spacy
nlp=spacy.load("en_core_web_sm")
def preprocess(text):
  doc=nlp(text)
  filtered_tokns=[token.lemma_ for token in doc if not token.is_stop or not token.is_punct]
  return " ".join(filtered_tokns)
```

```python
df["processed txt"]=df.Comment.apply(preprocess)
df.head(5)
```

```python
# train test splitt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    df['processed txt'],
    df.emotion_num,
    test_size=0.2,
    stratify=df.emotion_num)
```

```python
# making the pipeline
clf=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("rfc",RandomForestClassifier())
])

clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print(classification_report(y_test,y_predict))
```

```python
clf.score(x_test,y_test)
```
